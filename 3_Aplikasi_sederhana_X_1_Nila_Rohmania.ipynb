#Aplikasi Menghitung Volume Bangun Balok

```python
panjang = int(input("Masukkan panjang dari balok (cm): "))
lebar = int(input("Masukkan lebar dari balok (cm): "))
tinggi = int(input("Masukkan tinggi dari balok (cm): "))

volume = panjang * lebar * tinggi

print(f"\nVolume dari baloknya adalah {volume} cm^3.")
```
